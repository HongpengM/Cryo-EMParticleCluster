```python
sourcedata = readh5.mergeDataSet('../weiner', r'(.*?)weiner\.hdf')
```

```python
def plotamat(mat):
    fig = plt.figure()
    plt.imshow(mat, vmin=mat.min(), vmax=mat.max(), cmap=plt.cm.gray)
    plt.show()
```

```python
mask = np.zeros(data.shape[1:])
for i in range(mask.shape[0]):
    for j in range(mask.shape[1]):
        if (i - mask.shape[0] / 2)**2 + (j - mask.shape[1] / 2)**2 < (mask.shape[0]/2)**2:
            mask[i,j] = 1
plotamat(mask)
            

```
